```python
# Game Theory - The Python Equilibrium Assigment
# Catarina Jesus, Student ID: 26810706
# Python for Data Analysts: Methods & Tools - DAT 7466 - BMBAN1

#Full description of the game theory situation:

#The code is a game theory situation that simulates the decision-making process surrounding a country and the rest of the world to vaccinate against covid-19 and the implications of not doing it. 
#The game begins by asking the player if they want to play with the computer. There are 2 choices, “yes” or “no”, if the player answers something else, it will prompt the player again until a valid response is given. 
#If the player decides to play, the game will ask for a name of a country followed by a table with potential outcomes for that country and the rest of the world based on whether they choose.
#The game prompts the player to choose whether or not the country should adopt the vaccine and generates a random choice for the computer´s decision. The game uses a payoff matrix to give a final answer regarding the choices, allocating points, accordingly, printing out the results, and the points earned by each.
#If the player doesn’t decide to play, the computer it will play by itself, choosing randomly a country followed by a table with the potential outcomes for that country and the rest of the world based on whether they choose.
#The game will generate a random choice for both computers to choose whether or not the country should adopt the vaccine. The game uses a payoff matrix to give a final answer regarding the choices, allocating points, accordingly, printing out the results, and the points earned by each. Finally, the game will generate a winner after running 5 times. 
```

```python
#Import random package to generate random numbers. 

import random

#Ask if the player wants to play with the computer.

player = ""

#Condition if the player says something different that yes/no.

while player !="yes" and player!="no" :
    player = input("""Hello! Welcome to the most exciting game that you could ever play.
    Would you like to join me? (yes/no): """ ).lower()


#Condition if the player says yes.

if player == "yes":
    country_name = input("""Hello again. Are you prepared to have some fun while you learn more about covid-19 vaccination. To start I would like to choose a country: """).capitalize()
    
    print(f"\n{country_name}, as you know the Covid-19 pandemic was a big issue around the world.")
    
    print("""Each country has a choice to either take the vaccine or not, however, the decision can influence the rest of the population.  
    Imagine the following situation:""")
    
    #Table Game.
    
    import pandas as pd
    
    option_1 = ["Immunity can be achieved", f"{country_name} will be isolated from the world"]
    option_2 = [f"{country_name} will be isolated from the world", "The infection levels will rise"]
    
    options_list = [option_1, option_2]
    
    df = pd.DataFrame(options_list,  index = ["Take Vaccine", "Don´t take vaccine"], columns = ["Take Vaccine", "Don´t take vaccine"])
    
    print(df.to_string(index = ["Take Vaccine", "Don´t take vaccine"], columns = ["Take Vaccine", "Don´t take vaccine"], justify = 'center'))
 
    
    # Create a board game with the pay-off matrix.
    
    board_game = [[5, 2], [1, 0]]
    
    #Save board points and results in a dictionary. TV: Take Vaccine, DTV: Don´t Take Vaccine
    
    dictionary_results = {"(TV, TV)": (board_game[0][0], board_game[0][0]),
                         "(TV, DTV)": (board_game[0][1], board_game[1][0]),
                         "(DTV, TV)": (board_game[1][0], board_game[0][1]),
                         "(DTV, DTV)": (board_game[1][1], board_game[1][1])}
    
    # prompt the user to select what to do.
    
    user_choose = input("""
    What would you be the best solution for you? Would you advise the country to take the vaccine or don’t take it?
    Select:
    - TV: Take Vaccine
    - DTV: Don´t Take Vaccine
    """).upper()
    
    # Computer (Python) choses randomly or to take vaccine or don´t take vaccine.
    
    computer_choose = random.randint(0,1) # 0: take vaccine; 1: don´t take vaccine
    
    #Conditions for each scenario output.
    
    if user_choose == "TV" and computer_choose == 0:
        user_result = dictionary_results["(TV, TV)"][0]
        computer_result = dictionary_results["(TV, TV)"][1]
        print(f"""
    Both as us decided to take the vaccine, well done!
    You earned {user_result} points.
    I earned {computer_result}""")
    
    elif user_choose == "DTV" and computer_choose == 0:
        user_result = dictionary_results["(DTV, TV)"][0]
        computer_result = dictionary_results["(DTV, TV)"][1]
        print(f"""
    You decided to not take the vaccine, however you will be isolated from the rest of the world
    You earned {user_result} points.
    I earned {computer_result}""")

    elif user_choose == "TV" and computer_choose == 1:
        user_result = dictionary_results["(TV, DTV)"][0]
        computer_result = dictionary_results["(TV, DTV)"][1]
        print(f"""
    You decided to take the vaccine, however you will be isolated from the rest of the world
    You earned {user_result} points.
    I earned {computer_result}""")

    elif user_choose == "DTV" and computer_choose == 1:
        user_result = dictionary_results["(DTV, DTV)"][0]
        computer_result = dictionary_results["(DTV, DTV)"][1]
        print(f"""
    Both of us are not going to take the vaccine, we will increase the infection levels
    You earned {user_result} points.
    I earned {computer_result}""")
        
#Declaring a winner.

    if user_result > computer_result:
        print(f"""Amazing! You won the game, congratulations.""")

    elif user_result == computer_result:
        print(f"""Well we tie!""")

    elif user_result <= computer_result:
        print(f"""I beat you, congratulations to me!""")     

        
#Condition if the player says no - the computer (Pyhton) it will play with itself
    
else:
    
    #Computer (Python) will choose randomly a contry.
    
    country_name = ["Portugal", "Spain", "Mexico", "USA"]
    random_country = random.choice(country_name)
    
    computer_1=0
    computer_2=0
    computer_choose1 = "TV"
    
    #To run the program between the 2 computers (Python) 5 times
    
    for i in range(5): 
        
        print(f"\n{random_country}, as you know the Covid-19 pandemic was a big issue around the world.")
        
        print("""Each country has a choice to either take the vaccine or not, however, the decision can influence the rest of the population.  
        Imagine the following situation:""")
        
        #Table Game.
                
        import pandas as pd
        
        option_1 = ["Immunity can be achieved", f"{random_country} will be isolated from the world"]
        option_2 = [f"{random_country} will be isolated from the world", "The infection levels will rise"]
        
        options_list = [option_1, option_2]
        
        df = pd.DataFrame(options_list,  index = ["Take Vaccine", "Don´t take vaccine"], columns = ["Take Vaccine", "Don´t take vaccine"])
        
        print(df.to_string(index = ["Take Vaccine", "Don´t take vaccine"], columns = ["Take Vaccine", "Don´t take vaccine"], justify = 'center'))
        
        #Create a board game with the pay-off matrix.
        
        board_game = [[5, 2], [1, 0]]
        
        #Save board points and results in a dictionary. TV: Take Vaccine, DTV: Don´t Take Vaccine
        
        dictionary_results = {"(TV, TV)": (board_game[0][0], board_game[0][0]),
                              "(TV, DTV)": (board_game[0][1], board_game[1][0]),
                              "(DTV, TV)": (board_game[1][0], board_game[0][1]),
                              "(DTV, DTV)": (board_game[1][1], board_game[1][1])}
        
        #Both Computer (Python) chosee randomly or to take vaccine or don´t take vaccine.
        
        if computer_choose1 == "TV":
                computer_choose2="DTV"
        else:
            computer_choose2="DTV"
        computer_choose1 = random.choice(["TV", "DTV"])
        print("Computer 1 chose: ", computer_choose1)
        print("Computer 2 chose: ", computer_choose2)
        
        #Conditions for each scenario output.
        
        if computer_choose1 == 'TV' and computer_choose2 == 'TV':
            computer_result1 = dictionary_results["(TV, TV)"][0]
            computer_result2 = dictionary_results["(TV, TV)"][1]
            print(f"""
            Both as us decided to take the vaccine, well done!
            You earned {computer_result1} points.
            I earned {computer_result2}""")
            computer_1+=board_game[0][0]
            computer_2+=board_game[0][0]
        
        elif computer_choose1 == 'DTV' and computer_choose2 == 'TV':
            computer_result1 = dictionary_results["(DTV, TV)"][0]
            computer_result2 = dictionary_results["(DTV, TV)"][1]
            print(f"""
            You decided to not take the vaccine, however you will be isolated from the rest of the world
            You earned {computer_result1} points.
            I earned {computer_result2}""")
            computer_1+=board_game[1][0]
            computer_2+=board_game[0][1]
        
        elif computer_choose1 == 'TV' and computer_choose2 == 'DTV':
            computer_result1 = dictionary_results["(TV, DTV)"][0]
            computer_result2 = dictionary_results["(TV, DTV)"][1]
            print(f"""
            You decided to take the vaccine, however you will be isolated from the rest of the world
            You earned {computer_result1} points.
            I earned {computer_result2}""")
            computer_2+=board_game[1][0]
            computer_1+=board_game[0][1]
        
        elif computer_choose1 == 'DTV' and computer_choose2 == 'DTV':
            computer_result1 = dictionary_results["(DTV, DTV)"][0]
            computer_result2 = dictionary_results["(DTV, DTV)"][0]
            print(f"""
            Both of us are not going to take the vaccine, we will increase the infection levels
            You earned {computer_result1} points.
            I earned {computer_result2}""")
            computer_2+=board_game[1][1]
            computer_1+=board_game[1][1]
        
        #Declaring the winner between the two computers.
        
        if computer_1 > computer_2:
            print(f'Computer 1 scored {computer_1} points and Computer 2 scored {computer_2} \n Computer 1 won')
        
        elif computer_2 > computer_1:
            print(f'Computer 2 scored {computer_2} points and Computer 1 scored {computer_1} \n Computer 2 won')
        else:
            print('Well we tie!')
```
